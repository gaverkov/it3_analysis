```python
R = PolynomialRing(QQ,'x1,x2')
x1,x2 = R.gens()
x = vector(R.gens())

A = matrix(QQ,[(3,1),(1,2)])
#A = matrix(QQ, [(2,-1),(-1,2)])

pretty_print(A)
q = x*A*x
pretty_print(q)

CP = contour_plot(q,(x1,-1.5,1.5),(x2,-1.5,1.5),fill=False,contours=list(range(1,12)))

D, U = A.eigenmatrix_right()
eigenvalues = D.diagonal()
eigenvectors = [vector(u)/norm(u) for u in U.columns()]
ARROWS = sum([arrow(0*vec,vec/sqrt(val)) for val,vec in zip(eigenvalues,eigenvectors)])

show(CP+ARROWS)
```
