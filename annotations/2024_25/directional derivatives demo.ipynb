```python
var('x,y')
import numpy as np
f(x,y) = cos(y - x^2)+1
print("Die Funktion")
print(f)
opt = {'projection' : 'orthographic','color': 'yellow', 'opacity' : 0.8}
plotf = plot3d(f(x,y),(x,-1,1.5),(y,-1,1.5), **opt)
print("Der Gradient")
g = f.gradient()
print(g)
pt = vector((1.0,0.25))
print("An diesem Punkt werden wir die Richtungsableitungen auswerten")
print(pt)


T = np.linspace(0,6.3,num=10)
H = [0.5*vector([cos(t),sin(t)]) for t in T]
print("Hier sind die Richtungsvektoren, die wir behandeln")
for h in H:
    print(h)

vectors_along_surface = [list(pt + h) + [f(*pt)+g(*pt)*h] for h in H]
graphpt = list(pt) + [f(*pt)]

shadow_vectors = [ list(pt+h) + [0] for h in H]
shadow_graphpt = list(pt) + [0]


plot_arrows = sum([arrow3d (graphpt,d,color='blue') for d in vectors_along_surface])
plot_shadow_arrows = sum([arrow3d (shadow_graphpt,d,color='red') for d in shadow_vectors])
show(plotf + plot_arrows + plot_shadow_arrows)
```
