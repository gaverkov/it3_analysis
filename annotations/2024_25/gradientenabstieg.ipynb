```python
var('x,y')
f(x,y) = sum([-cos(a*x+b*y+c) for a,b,c in [(1,1,1),(-1,1,0),(1,1,1),(0,-1,2)]])
f(x,y) = (x-2.0)^2 + (y-1.0)^2 + (2*x^2 + y^2 + 3)^2
print(f)

PLOT_F = plot3d(f(x,y),(x,-5,5),(y,-5,5),projection='orthographic',color='yellow',opacity=1)

CONT_PLOT_F = contour_plot(f(x,y),(x,-5,5),(y,-5,5),fill=False,cmap='hsv')

def gradient_descent_process(f,start,step,max_iter=200):
    g = f.gradient()
    current_position = vector(RDF,start)
    steps = [tuple(current_position)]
    for i in range(max_iter):
        current_val = RDF(f(*current_position))
        next_position = current_position-step*vector(RDF,g(*current_position))
        next_val = f(*next_position)
        if next_val >= current_val:
            return steps
        current_position = next_position
        steps.append(tuple(current_position))
    return steps

descents = []
descents_3d = []
for reps in range(10):
    descent = gradient_descent_process(f,(-3.0+6.0*random(),-3.0+6.0*random()),0.1)
    descents.append(descent)
    descents_3d.append([list(pt)+[f(*pt)+0.1] for pt in descent])

LINES = sum([line(descent,color='black',thickness=2) for descent in descents])
LINES_3D = sum([line(descent_3d,color='black',thickness=4) for descent_3d in descents_3d])
```

```python
show(CONT_PLOT_F+LINES)
show(PLOT_F + LINES_3D)
```
