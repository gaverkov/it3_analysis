```python
var('x')
def newton_step(f,p):
    """
        p ist der aktuelle Punkt
        wir sind auf der Suche nach der nächsten 
        Approximation von f(x) = 0.
        Das ist unsere Rückgabe (die nächste Approximation)
    """
    return p - f.diff()(p)^(-1)*f(p)
```

```python
f(x) = 1.0 * (e^x + x^2)
```

```python
plot(f(x),(x,-1,1)) + point( (q,f(q)), size =50)
```

```python
q = 1.0
g = f.diff()
for i in range(3):
    q = newton_step(g,q)
```
