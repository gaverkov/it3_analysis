```python
var('x,y')
f(x,y) = x^2 + 4*y^2 - 1 + 1.5*x + 0.7*y + 0.3*(2*x-y)^2
g(x,y) = x^2 + y^2 - 1 - 0.2*x + 0.5*y
```

```python
pf  = implicit_plot(f,(x,-2,2),(y,-2,2),color='blue')
pg  = implicit_plot(g,(x,-2,2),(y,-2,2),color='red')

def F(x,y):
    return vector([f(x,y),g(x,y)])

Fstrich = matrix([f.gradient(),g.gradient()])
```

```python
p = vector([-2.0,2.0])
#p = vector([42,17])
```

```python
def next_position(p):
    return p - Fstrich(*p).inverse()*F(*p)
```

```python
q = p
verlauf = [q]
for i in range(5):
    q = next_position(q)
    verlauf.append(q)
```
